## 1.Content Based Recommendation System

```python
# TF_IDF Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(stop_words="english",max_df=.8,min_df=5,ngram_range=(1,2))
tfidf_matrix=vectorizer.fit_transform(df["headline"])
tfidf_matrix.toarray()
```

```python
cosin_sim=cosine_similarity(tfidf_matrix,tfidf_matrix)
cosin_sim
```

## Recommend Function

```python
pd.set_option('display.max_colwidth', None)
```

```python
title="The investment stratergies in 2025"
import re
def content_base_rec(title,n_rows=5):
    title=re.sub(r"\W+"," ",title) # removes special characters
    title=title.lower().strip()
    title_vc=vectorizer.transform([title])
    cosine_scores=cosine_similarity(title_vc,tfidf_matrix).flatten()
    indices=cosine_scores.argsort()[::-1][:5]
    return df.loc[indices,["headline"]]
content_base_rec("education is free")
```
